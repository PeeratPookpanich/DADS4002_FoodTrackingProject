# ข้อมูลภาพรวมของ user

## Query ข้อมูลภาพรวมของ user

```python
import mysql.connector
from getpass import getpass

user = input("User name: ")
pwd = getpass('Password')

my_db,my_cursor = None, None
try:
    my_db = mysql.connector.connect(host='localhost',
                                   user=user,
                                   password=pwd,
                                   database='dads4002_final')
    print("Successfully connected to the database")
    
    # 1.Create a cursor for this database connection.
    my_cursor = my_db.cursor() 
  
    # SQL command 
    # Create strings of MySQL commands
    sql_commands = """
                    SELECT count(log_id) as data_number,
                    min(date) as start_date,
                    max(date) as end_date
                    FROM foodlogs;
                    -- -----------------------------------
                    SELECT count(user_id) as subject_number,
                    avg(timestampdiff(year,birthday,now())) as Avg_Age,
                    max(timestampdiff(year,birthday,now())) as MAX_Age,
                    min(timestampdiff(year,birthday,now())) as MIN_Age,
                    avg(init_weight) as Avg_Weight,
                    max(init_weight) as MAX_Weight,
                    min(init_weight) as MIN_Weight
                    FROM users;
                    -- -----------------------------------
                    SELECT gender, count(user_id) as number_byGender
                    FROM users
                    group by gender;"""
    
    # Execute multiple SQL commands and iterate for each query result
    # Iterate on the returned value of execute() in order to access results from each query separately.
    for i,result in enumerate(my_cursor.execute(sql_commands, multi = True)):
        #print(f"\n===== Query result no.{i+1} =====")
        if result.with_rows:
            #print(result.fetchall())
            # loop through the rows
            for row in result.fetchall():
                print()
                if i == 1:
                    #print("======== ภาพรวมของข้อมูลทั้งหมด ========\n")
                    print(f' จำนวนผู้ใช้งาน: {row[0]} คน \n อายุเฉลี่ย(ปี): {row[1]:.0f} ซึ่งอยู่ในช่วง {row[3]} ถึง {row[2]}\n น้ำหนักเฉลี่ย(kg): {row[4]:.0f} ซึ่งอยู่ในช่วง {row[6]} ถึง {row[5]}')
                if i == 0:
                    print("======== ภาพรวมของข้อมูลทั้งหมด ========\n")
                    print(f' จำนวนข้อมูลทั้หมด: {row[0]} rows \n เริ่มเก็บเมื่อวันที่ {str(row[1])} ถึง {str(row[2])}')
                if i == 2:
                    print(f' จำนวน {row[0]}: {row[1]} คน' )
                  
        else:
            print(f"Number of rows affected by statement '{result.statement}':{result.rowcount}")
                
except Exception as e:
    print("Something went wrong while attempting to connect the database.")
    if my_db is not None: 
        my_db.rollback()
    print(f">>>Error{e}")
finally:
    if my_cursor is not None:
        my_cursor.close()
    if my_db is not None:
        my_db.close()
        
    print("\n Database closed")
```

# ข้อมูลภาพรวมของอาหารแต่ละประเภท

## Query ข้อมูลอาหารแต่ละประเภท

```python
import mysql.connector
from getpass import getpass

user = input("User name: ")
pwd = getpass('Password')

my_db,my_cursor = None, None
try:
    my_db = mysql.connector.connect(host='localhost',
                                   user=user,
                                   password=pwd,
                                   database='dads4002_final')
    print("Successfully connected to the database")
    
    # 1.Create a cursor for this database connection.
    my_cursor = my_db.cursor() 
  
    # SQL command 
    # Create strings of MySQL commands
    sql_commands = """
                    SELECT foodtype.type_name,
                    count(foodlogs.type_id) as count_food_type,
                    (count(foodlogs.type_id)/max(foodlogs.log_id))*100 as percent_food_type
                    FROM foodlogs
                        join foodtype
                        on foodlogs.type_id = foodtype.type_id
                    group by foodlogs.type_id
                    order by count_food_type desc;
                    -- --------------------------------
                    SELECT food_national,
                    count(foodlogs.food_national) as count_food_national,
                    (count(foodlogs.food_national)/max(foodlogs.log_id))*100 as percent_food_national
                    FROM foodlogs
                    group by foodlogs.food_national
                    order by count_food_national desc;
                    -- --------------------------------
                    SELECT foodlogs.type_id,
                    foodtype.type_name,
                    foodlogs.food_group,
                    count(foodlogs.food_group) as count_food_group,
                    (count(foodlogs.food_group)/( select count(foodlogs.food_group)
                                                        FROM foodlogs
                                                        where foodlogs.type_id = "T00002"))*100 as percent_food_group
                    FROM foodlogs
                        join foodtype
                        on foodlogs.type_id = foodtype.type_id
                    where foodlogs.type_id = "T00002"
                    group by foodlogs.food_group
                    order by count_food_group desc;
                    -- --------------------------------
                    SELECT foodlogs.type_id,
                    foodtype.type_name,
                    foodlogs.food_group,
                    count(foodlogs.food_group) as count_food_group,
                    (count(foodlogs.food_group)/( select count(foodlogs.food_group)
                                                        FROM foodlogs
                                                        where foodlogs.type_id = "T00001"))*100 as percent_food_group
                    FROM foodlogs
                        join foodtype
                        on foodlogs.type_id = foodtype.type_id
                    where foodlogs.type_id = "T00001"
                    group by foodlogs.food_group
                    order by count_food_group desc;
                    -- --------------------------------
                    SELECT foodlogs.type_id,
                    foodtype.type_name,
                    foodlogs.food_group,
                    count(foodlogs.food_group) as count_food_group,
                    (count(foodlogs.food_group)/( select count(foodlogs.food_group)
                                                        FROM foodlogs
                                                        where foodlogs.type_id not in ( "T00001","T00002")))*100 as percent_food_group
                    FROM foodlogs
                        join foodtype
                        on foodlogs.type_id = foodtype.type_id
                    where foodlogs.type_id not in ( "T00001","T00002")
                    group by foodlogs.food_group
                    order by count_food_group desc;"""
    
    # Execute multiple SQL commands and iterate for each query result
    # Iterate on the returned value of execute() in order to access results from each query separately.
    for i,result in enumerate(my_cursor.execute(sql_commands, multi = True)):
        if i == 0:
            result11 = result.fetchall()
        if i == 1:
            result12 = result.fetchall()
        if i == 2:
            result13 = result.fetchall()
        if i == 3:
            result14 = result.fetchall()
        if i == 4:
            result15 = result.fetchall()
        
        #else:
            #print(f"Number of rows affected by statement '{result.statement}':{result.rowcount}")
                
except Exception as e:
    print("Something went wrong while attempting to connect the database.")
    if my_db is not None: 
        my_db.rollback()
    print(f">>>Error{e}")
finally:
    if my_cursor is not None:
        my_cursor.close()
    if my_db is not None:
        my_db.close()  
    print("\nDatabase closed")
```

## Query ข้อมูลพฤติกรรมของ user

```python
user = input("User name: ")
pwd = getpass('Password')

my_db,my_cursor = None, None
try:
    my_db = mysql.connector.connect(host='localhost',
                                   user=user,
                                   password=pwd,
                                   database='dads4002_final')
    print("Successfully connected to the database")
    
    # 1.Create a cursor for this database connection.
    my_cursor = my_db.cursor() 
  
    # SQL command 
    # Create strings of MySQL commands
    sql_commands = """
                    -- sum price
                    SELECT foodlogs.user_id,
                    WEEK(foodlogs.`date`,0) as Week_number,
                    sum(foodlogs.food_price) as price_perWeek
                    FROM foodlogs
                    group by WEEK(foodlogs.`date`,0),user_id
                    order by WEEK(foodlogs.`date`,0), price_perWeek desc;
                    -- ---------------------------------
                    SELECT user_id, Week_number, avg(daily_cals ) as Avgcalories_perWeek
                    FROM (select user_id,`date`, WEEK(`date`,0) as Week_number, sum(calories) as daily_cals from foodlogs group by user_id,`date`) as aa
                    group by Week_number,user_id
                    order by Week_number, Avgcalories_perWeek desc;
                    -- --------------------------------
                    SELECT user_id, WEEK(`date`,0) as Week_number,
                        avg(daily_step) as Avgstep_perWeek
                    FROM usertracking
                    group by Week_number,user_id
                    order by Week_number, Avgstep_perWeek desc;
                    -- --------------------------------
                    -- avg sleep
                    SELECT user_id, WEEK(`date`,0) as Week_number,
                        avg(daily_sleep) as Avgsleep_perWeek
                    FROM usertracking
                    group by Week_number,user_id
                    order by Week_number, Avgsleep_perWeek desc;
                    -- --------------------------------
                    SELECT user_id, WEEK(`date`,0) as Week_number,
                        avg(daily_drink) as Avgdrink_perWeek
                    FROM usertracking
                    group by Week_number,user_id
                    order by Week_number, Avgdrink_perWeek desc;
                    -- --------------------------------
                    select distinct hour(time), count(hour(time))
                    from foodlogs
                    group by hour(time)
                    order by hour(time);
                    """
    
    # Execute multiple SQL commands and iterate for each query result
    # Iterate on the returned value of execute() in order to access results from each query separately.
    for i,result in enumerate(my_cursor.execute(sql_commands, multi = True)):
        if i == 0:
            result21 = result.fetchall()
        if i == 1:
            result22 = result.fetchall()
        if i == 2:
            result23 = result.fetchall()
        if i == 3:
            result24 = result.fetchall()
        if i == 4:
            result25 = result.fetchall()
        if i == 5:
            result26 = result.fetchall()
        
        #else:
            #print(f"Number of rows affected by statement '{result.statement}':{result.rowcount}")
                
except Exception as e:
    print("Something went wrong while attempting to connect the database.")
    if my_db is not None: 
        my_db.rollback()
    print(f">>>Error{e}")
finally:
    if my_cursor is not None:
        my_cursor.close()
    if my_db is not None:
        my_db.close()
        
    print("\n Database closed")
```
